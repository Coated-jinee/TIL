# 2024년 11월 5일 화요일

## 허깅페이스(Hugging Face)
AI 커뮤니티와 모델 있음

### 허깅페이스의 특징
1. Transformers 라이브러리 제공 
- BERT, GPT-3 같은 최신 NLP 모델을 쉽게 사용할 수 있음
2. 모델 허브 
- 클릭 몇번으로 모델을 가져다 쓸 수 있음
3. 커뮤니티 중심
- 오픈소스 커뮤니티 중심 운영

### 허깅페이스 장점
1. 쉬운 접근성
2. 광범위한 모델 선택
3. 오픈소스
4. 강력한 커뮤니티 지원

### 허깅페이스 단점
리소스 요구량
복잡한 초기 설정
특화된 모델
- NLP 외의 다른 AI 분야에서는 상대적으로 모델의 수가 적음

### 허깅페이스 활용
1. 소설 쓰기
2. 챗봇 만들기
