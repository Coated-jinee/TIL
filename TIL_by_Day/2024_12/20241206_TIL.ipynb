## Docker와 Container
### Docker란?
- Linux container를 기반으로 만든 OS 레벨 가상화 구현을 도와주는 프로그램
VM 이라는 (이건 써봄. Oracle.. )
코드 밀키트 (OS 위에 라이브러리, 개발용 SDK, 필요한 프로그램, 실행할 코드 다 얹을 수 있음)
도커를 쓰면 버튼 딸깍 눌러서 사용 가능해짐
이걸 사용하는 가상 컴퓨터를 컨테이너라고 부름

### Docker 만드는법
- IDE에서 Dockerfile을 하나 만든다. 
- 어떤 OS 쓸건지, 어떤 코드랑 라이브러리 쓸건지, 어떤 SDK 쓸건지, 어떤 코드를 어떻게 실행할건지 쓰고 명령어 한줄로 밀키트로 포장
- 도커 가서 클릭 한번으로 밀키트 실행
- 빌드 배포 테스트가 Docker 덕에 초등학생도 할 수 있게 됨

### Docker 
- Docker는 기존의 리눅스 가상화 환경(Linux Containers, LXC)을 보다 쉽게 사용할 수 있도록 만든 도구임
- 원래 리눅스에서 컨테이너를 띄우는 것은 가능했지만, Docker는 이를 일반 사용자도 편리하게 활용할 수 있도록 추상화하고, 관리 도구와 API를 제공한다.
- Docker 컨테이너의 특징 중 하나는 **휘발성 데이터**라는 것. 컨테이너 내부의 데이터는 컨테이너가 종료되면 기본적으로 사라진다. 이를 방지하기 위해서는 볼륨(volume)이나 바인드 마운트(bind mount)를 사용해 데이터를 지속적으로 저장할 수 있음

## 특강: Docker 보충
### Docker와 CI/CD 프로세스
- **CI (Continuous Intergration):** 개발자들이 코드를 브랜치에서 작성하고 PR(Pull Request)을 통해 코드를 병합(Merge)하면 자동으로 테스트와 검증이 진행됨
- **CD (Continuous Deployment/Delivery):** 테스트를 통과한 코드가 자동으로 사용자들에게 배포됨

#### CI/CD 기본 단계
1. 코드를 작성하고 GitHub와 같은 버전 관리 시스템에 푸시
2. Python과 같은 언어에서는 빌드 과정이 필요하지 않을 수 있지만, CI/CD 툴이 코드 리포지토리를 가져옴
3. 가져온 코드에 대해 테스트 및 검증을 수행
4. 검증이 완료되면 배포가 이루어짐

#### Q&A
- Q: CI/CD에서 주로 테스트 코드 자동화가 목적일까요?
- A: 일부 그렇다. 하지만 CI/CD는 테스트뿐 아니라 코드 배포를 효율적이고 지속적으로 수행하는 것이 주 목적이다.

#### 전체적인 프로세스 요약
- CI/CD를 통해 지속적인 검증과 배포를 반복하는 것이 핵심
- 전체 프레임(예: 프로젝트 구조)을 미리 설계하고, 이를 CI/CD로 검증 및 배포하며, 문제가 있으면 빠르게 수정해 다시 배포
- 일부 기능만 구현하고 시장에 빠르게 배포하여 피드백을 받아 다시 개발하는 방식이 트렌드로 자리 잡고있음

### Docker의 등장과 역할
#### Docker의 개념
- Docker는 2014년에 처음 소개되어 2016~2017년경 상용화 되었음 
- 동일한 환경에서 테스트와 배포를 가능하게 해주는 도구로, 개발 환경과 운영 환경 간의 불일치를 줄이는 데 큰 도움을 준다.
#### Docker를 사용하는 이유
1. 빌드 및 테스트를 보다 간단하고 일관되게 수행할 수 있다.
2. 애플리케이션의 실행 환경을 컨테이너로 묶어 운영 환경에서도 동일하게 배포할 수 있다.
3. CI/CD 프로세스와의 통합이 용이하며, 지속적인 배포의 효율성을 높인다.
