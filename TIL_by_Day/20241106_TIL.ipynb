## CLT(Central limit Theorem)
- 제발 이걸 왜 공부한건지 묻지 말아줘요. 나도 궁금하니까 ㅠㅋㅋ
- 확실한건 누군가 시켰음..

중심극한정리는 여러 표본의 평균값 분포가 모집단의 분포 형태에 상관없이 점차 정규분포를 따르게 된다는 개념이다. 중요한 요점은 표본의 크기가 커질수록 평균값 분포가 정규분포에 가까워진다는 것이다. 이 원리를 통해, 우리는 실제 모집단을 직접 분석하지 않고도 표본을 통해 모집단의 특성을 추정할 수 있다. 예를 들어, 생물학 실험에서 특정 유전자 발현 수준을 여러번 측정해 평균을 구하게 되면, 이 평균값들이 모여 정규분포를 이루게 된다. 이를 통해 모집단의 평균과 변동성을 추정할 수 있게 된다. 

정규분포에 가까워진다는 것의 의미는, 표본 평균의 분포가 예측 가능한 패턴을 따른다는 것이다. 이 점은 통계 분석과 추론에서 매우 중요한 의미를 가지고있다.

정규분포는 잘 정의된 특성을 가지고 있어서, 표본평균의 분포가 정규분포를 따르면 통계적으로 모집단의 특성을 추정하고 예측하기 쉬워진다. 이를 통해 얻는 이점은 크게 두 가지 이다: 

1. **신뢰구간 설정이 용이해진다:** 표본 평균이, 정규분포를 따른다면, 표본을 통해 모집단 평균의 범위를 쉽게 추정할 수 있다. 예를 들어, 특정 신뢰구간 내에 모집단 평균이 위치할 확률을 계산할 수 있게 된다.
2. **가설 검정에 활용할 수 있다:** 정규분포의 특성을 이용해 특정 가설이 맞는지를 판단할 수 있다. 예를 들어, 두 집단 간의 평균 차이를 비교하거나 실험 결과가 우연히 발생했을 가능성을 평가하는 데 정규분포가 활용된다.

쉽게 말해, 정규분포가 예측 가능성을 부여하기 때문에 통계적 결론을 더 신뢰할 수 있게 된다. 

CLT 한줄 요약
CLT는 많은 표본을 뽑아서 각각의 평균을 구하면, 이 평균값들이 모여 정규분포를 따른다는 의미다. 

### 정규분포
**정규분포는 종 모양(대칭형)분포인데, 왜 이름이 정규분포인가?**
- 정규분포(Normal Distributior): 이 이름은 영어에서 유래된 걸 봐야 이해가 쉬움. **"정규분포"**라는 이름은 **"정상적(Normal)"**이라는 의미에서 유래했다. 이 분포는 자연 현상, 생물학적 측정치(예: 키, 체중), 시험 성적 등에서 가장 흔하게 나타나는 **"정상적" 분포 패턴**이기 때문에 붙여진 이름이다. 

수학자들이 다양한 데이터들을 분석하면서 많은 경우에 평균 주변에 데이터가 몰리고 양쪽으로 점차 줄어드는 대칭형 패턴을 발견했다. 그래서 이 대칭형 패턴을 "정상적(Normal)"이라 보고, Normal distribution이라고 이름 붙였다.

### 정규분포 외의 분포들
정규분포 외에도 자주 사용하는 다양한 분포들이 있다. 대표적인 분포로는 다음과 같은 것들이 있다. 
1. 이항분포(Binomial Distribition)
- **"bi-" (두 개의)** + **"nomial" (용어 또는 항)**이 합쳐진 단어이다. 즉, 두 가지 결과를 가진 상황을 의미한다.
- 이항분포는 두 가지 결과가 있는 실험에서 나타나는 분포를 나타내는데, 예를 들어 "성공/실패", "참/거짓", 또는 "앞면/뒷면" 같은 두 개의 가능한 결과가 있을 때 사용된다.

2. 포아송 분포(Poisson Distribution)
-  단위 시간이나 공간 내에서 특정 사건이 발생할 확률을 나타낸다. 예를 들어, 특정 시간 동안 병원에 도착하는 환자의 수나, 특정 구간에서 발생하는 교통사고 건수 등을 분석할 때 사용된다. 

3. 카이제곱 분포(Chi-Square Distribution)
- 주로 데이터의 분산을 비교하거나 두 집단 간의 독립성을 검정할 때 사용된다. 예를 들어, 특정 질병이 남성과 여성에게 동등하게 나타나는지 확인할 때 유용하다.

4. t-분포(Student's t-Distribution)
- 표본의 크기가 작을 때, 특히 표본평균이 모집단 평균을 얼마나 잘 추정하는지를 평가할 때 사용된다. 작은 표본에서 정규분포 대신 자주 사용되는 분포이다.

5. 지수분포(Exponential Distribution)
- 특정 사건이 발생하기까지의 시간을 모델링할 때 사용된다. 예를 들어, 기계의 고장 발생까지의 시간이나 고객이 도착하는 간격 등을 분석할 때 유용하다. 

정리를 하면서도 혼란스럽다. 통계검정과 확률분포가 같이 쓰이는데 용어들이 너무 거부감들어서 학습이 안됨.
알고있던것도 다 엉켜서 전부 모르는 것처럼 되는중.
정규분포, 이항분포, 포아송분포, 카이제곱분포, t-분포, 지수분포 모두 **확률분포**이다.
다른 분포도 있지만 일단 확률분포만이라도 알자

### 통계검정과 확률분포의 개념분리하기
1. 확률분포(Probability Distribution)는 데이터를 설명하고 모델링하는 "기반"이다. 어떤 데이터가 특정 분포를 따른다고 가정하면, 그 분포의 특성을 이용해 확률 계산이나 예측을 할 수 있다. 예를 들어, 정규분포는 평균 주변에 데이터가 몰려있는 경우 사용하고, 포아송분포는 특정 시간이나 공간 내 드문 사건 발생에 사용한다.

2. 통계검정은 가설을 검증하는 방법이다. 데이터를 분석하여, 우리가 가진 가설이 참인지 아니면 우연에 의한 것인지 판단하는 데 사용된다. 예를 들어:
- t-test는 두 그룹 간 평균의 차이가 유의한지 확인하는 데 사용한다.
- Chi-squre test는 범주형 변수 간의 독립성이나 적합성을 검증할 때 사용한다.
- ANOVA는 세 개 이상의 그룹 간 평균 차이를 검증하는 데 사용한다.

#### 둘이 함께 사용되는 방식
- 분포는 검정의 이론적 기반이 되는 경우가 많다. 예를 들어 t-test는 데이터가 정규분포를 따를 때 유효한 검정이다. 이처럼 검정을 하기 전에 데이터가 특정 분포를 따른다고 가정하거나 확인하는 과정이 포함될 수 있다.
- 예를 들어, 카이제곱 검정은 카이제곱 분포를 기반으로 결과의 유의상을 판단하고, ANOVA의 F-분포나 t-test의 t-분포도 같은 방식이다. 

따라서, 분포는 데이터의 "형태"나 "특성"을 나타내는 기반이고, **통계검정은 이 데이터를 바탕으로 가설을 검증하는 도구**라고 이해하면 좀 더 분명해질 수 있다. 

### 회귀(Regression)

#### 왜 이름이 회귀인가?
- 회귀(Regression) 라는 용어는 과거로 돌아가거나 원래 상태로 되돌아가는 의미를 가지고있다. 이 용어는 통계학자 프랜시스 골턴(Francis Calton)이 처음 사용했는데, 그의 연구에서 유래된 것이다.

골턴은 부모와 자식의 키 사이의 관계를 연구하면서, **부모의 키가 매우 크거나 작아도 자식의 키는 평균에 가까워지는 경향**이 있음을 발견하였다. 이를 그는 "Regression toward the Mean." 즉 **"평균으로의 회귀"**라고 불렀다. 극단적인 값을 가진 데이터가 다시 평균 쪽으로 돌아가려는 경향이 있는 것을 표현한 것이다. 

그래서 회귀 분석에서는 입력 변수들이 종속 변수에 어떤 영향을 미치는지를 찾고, 이를 통해 예측 값을 평균에 가까운 값으로 "회귀"시킨다는 의미로 "회귀(Regrssion)"라는 용어가 쓰이게 되었다.

#### TMI (진짜 투머치 인포메이션)
#### 프랜시스 골턴은 어떻게 부모의 키가 매우 크거나 작아도 자식의 키는 평균에 가까워지는 경향이 있음을 발견했을까? 
프랜시스 골턴은 19세기 후반에 부모와 자식간의 키 관계를 연구하며, 키가 유전적으로 물려받는 특성임을 밝혀내고자 했다. 그는 다양한 가족 데이터를 수집하여 부모의 키와 자식의 키 사이의 관계를 분석했다.

#### 구체적인 연구 과정과 발견
1. 데이터 수집: 골턴은 여러 가족의 키 데이터를 수집했는데, 이 데이터에는 부모(특히 아버지와 어머니의 평균 키)와 자식의 키가 포함되어있었다.
2. 데이터 분석: 부모의 키가 평균보다 크거나 작은 경우에 자식의 키는 평균 쪽으로 다소 수렴하는 경향이 있음을 관찰하였다. 예를 들어, 부모가 평균보다 매우 큰 경우라도 자식의 키는 부모보다는 작고, 반대로 부모가 매우 작은 경우 자식의 키는 부모보다는 큰 경향이 있었다. 
3. 평균으로의 회귀: 그는 이런 현상을 "Regression toward the Mean"이라고 명명했다. 이 개념은 극단적인 키를 가진 부모일수록 자식의 키가 모집단의 평균에 가까워지는 경향이 있다는 뜻이다. 골턴은 이를 통해, 부모의 키가 자식의 키에 영향을 주지만 완벽하게 일치하지 않으며, 자연적인 변동성으로 인해 자식의 키가 평균으로 돌아가려는 경향이 생긴다고 결론을 내렸다.

#### 통계적 원리와 회귀 분석
이 현상은 단순히 키뿐만 아니라 자연에서 흔히 나타나는 패턴이다. 극단적인 데이터(예: 높은 시험 점수, 아주 큰 키 등)가 후속 세대나 추가 측정에서 평균에 가까워지는 경향을 보이는 것은 자연적인 변동성과 평균으로의 회귀 경향 때문이다. 골턴의 이 발견은 오늘날의 회귀 분석의 기초가 되었고, 극단적인 값이 다시 평균에 가까워지는 현상을 설명하는 중요한 개념으로 자리 잡게 되었다. 

#### 근데 일반적으로 부모 쪽의 한쪽이 크면(아빠) 자식(아들)의 키가 대부분 크던데?
일반적으로 부모 중 한쪽이 큰 경우 자식도 큰 키를 물려받을 가능성이 크다. 유전적으로 키는 부모의 영향을 많이 받기 때문인데, 골턴의 연구에서 말하는 "평균으로의 회귀"는 이런 일반적인 유전적 경향과는 약간 다르게, 극단적인 부모의 키가 완전히 자식에게 전달되지 않고 평균에 가까워지려는 경향을 말하는 것이다. 

즉, 부모가 평균보다 크다면 자식도 큰 경향이 있지만, 부모만큼 극단적으로 크지 않고 평균에 약간 가까워질 가능성이 크다는 의미이다. 반대로 부모가 평균보다 작다면 자식도 작은 경향을 가지지만, **부모보다는 약간 커질 가능성이 있다는것**이다.

#### 회귀는 왜 여러 개념들이 생겼나?
회귀 개념이 처음 나온 후 여러가지 발전과 응용이 이루어지면서 다양한 회귀 개념들이 생겨났다. 초기에는 단순히 변수 간의 관계를 설명하기 위한 선형 회귀로 시작했지만, 데이터를 설명하고 예측하는 과정에서 다양한 형태의 데이터와 상황이 생기면서 회귀에 점점 "살이 붙은" 셈 이다.

#### 회귀 개념의 발전 과정
1. **단순 선형 회귀:** 초기에, 회귀 분석은 한 개의 독립 변수와 종속 변수 간의 직선 관계를 설명하는 데 사용되었다. 이는 두 변수 간의 상관관계를 설명하는 가장 기본적인 형태의 회귀였다.

2. **다중 선형 회귀:** 이후, 여러 변수들이 종속 변수에 영향을 미치는 상황이 많아지면서 다중 선형 회귀가 등장하였다. 예를 들어, 집값을 예측할 때 방 수, 위치, 면적 등 여러 변수를 고려해야 하는 경우처럼, 다중 회귀는 여러 변수의 영향을 동시에 분석하는 방법이 필요해지면서 발전 한 것 이다. 

3. **비선형 회귀:** 데이터가 직선이 아닌 곡선 형태로 나타나는 경우, 선형 회귀만으로는 설명하기 어려운 문제들이 발생하였다. 이에 따라 비선형 회귀가 도입되어, 데이터에 맞는 곡선 형태의 모델을 사용할 수 있게 되었다. 

4. **로지스틱 회귀와 그 외의 회귀 변형:** 회귀가 예측뿐만 아니라 분류 문제에도 적용되면서, 확률 기반의 로지스틱 회귀(Logistic Regression) 같은 기법이 탄생하였다. 로지스틱 회귀는 이진 분류 문제에 사용되며, 선형 회귀의 확장 개념으로 확률을 예측하는 방법이다.

5. **정규화 및 규제 회귀(Ridge, Lasso 회귀 등):** 데이터가 과적합되는 것을 방지하기 위해 정규화 및 규제 기법이 추가된 회귀 모델들이 개발되었다. 이 기법들은 복잡한 데이터를 다루면서도 과적합 문제를 완화시키기 위해 등장하였다. 

#### 다양한 회귀 개념의 필요성
현실에서는 데이터가 단순하지 않고, 비선형적이거나 여러 변수 간의 복잡한 상호작용을 포함하는 경우가 많다. 그래서 기본 회귀 모델을 기반으로 점점 다양한 변형과 추가 기능이 붙게 된 것이다. 각 회귀 모델은 데이터의 특성과 목적에 맞게 적용될 수 있도록 발전한 결과이다.

따라서 회귀는 하나의 기본 개념에서 시작되어, 실제 데이터와 문제 상황에 적합하게 다양한 모델들이 생겨난 것이라고 볼 수 있다.
