#### 1. 로지스틱 회귀의 활용
##### 1. 이진 분류(Binary Classfication)
- **이진 분류**란 목표 변수가 두 개의 클래스(0 또는 1, 예 또는 아니오 등)중 하나로 분류되는 문제이다.
- **로지스틱 회귀**는 입력 데이터(feature)와 목표 변수(target)사이의 관계를 모델링하여, 특정 클래스에 속할 확률을 계산한다. 계산된 확률을 기준으로, 특정 기준(일반적으로 0.5)를 넘어가면 해당 클래스로 예측한다.
- **예시**:
    - 타이타닉 데이터셋에서 생존자 예측 (생존:1 사망:0)
    - 이메일 스팸 분류 (스팸:1, 정상: 0)
    - 질병 진단 (질병있음:1, 질병 없음:0)

##### 2. 다중 클래스 분류(Multicalss Classfication)
- **다중 클래스 분류는** 목표 변수가 셋 이상의 클래스로 구성된 문제를 의미한다. 예를 들어, 분류해야 할 클래스가 'A', 'B', 'C' 세 가지인 경우가 이에 해당된다.
- 로지스틱 회귀는 **소프트맥스 함수**를 사용하여 다중 클래스 분류를 수행한다. 이를 **다항 로지스틱 회귀(Multinomial Logistic Regression)**라고 한다.
- **예시**:
    - 손글씨 숫자 인식 (0 ~ 9)
    - 붓꽃 품종 분류 (Setosa, Versicolor, Virginica)

##### **로지스틱 회귀의 작동 원리**
- 선형 회귀와는 달리, 로지스틱 회귀는 선형 방정식을 기반으로 한 예측 결과를 **시그모이드 함수(sigmoid function)**에 통과시켜 **확률값**으로 변환한다.
- 시그모이드 함수의 값은 0과 1사이의 범위를 가지며, 이 값이 클래스의 확률을 나타낸다. 예를 들어, 확률이 0.5 이상이면 클래스 '1'로, 0.5 미만이면 클래스 '0'으로 분류한다.

##### **로지스틱 회귀의 장점과 특징
- **단순하고 빠른 계산**: 수학적으로 단순하고 계산이 빠르며, 해석하기 쉽다.
- **확률 기반 예측**: 특정 클래스에 속할 확률을 제공하므로, 모델의 신뢰도를 평가할 수 있다.
- **선형 분류기**: 데이터가 선형적으로 구분 가능한 경우 효과적으로 사용할 수 있다.
