### Virtual environment: 가상환경

파이썬을 공부하다 보면, **'가상환경'** 이라는 말이 종종 나온다.

가상환경이 필요한 이유는 앞으로 프로젝트를 진행할 때, 특정 프로젝트 별로 독립적인 패키지와 설정을 관리할 수 있는 환경을 의미한다. 

가상환경을 사용하는 주요 이유는 다음과 같다.

1. **패키지 충돌 방지**
- 여러 프로젝트에서 서로 다른 버전의 Python 패키지를 사용할 수 있다. 예를 들어, 하나의 프로젝트에서는 pandas 버전 1.2를 사용하고, 다른 프로젝트에서는 pandas 버전 1.3을 사용할 수 있는데, 이를 하나의 환경에 모두 설치하면 충돌이 발생할 수 있다.
2. **프로젝트 간 종속성 관리**
- 각 프로젝트에 필요한 라이브러리와 패키지를 관리하고, 특정 프로젝트에만 필요한 패키지다 다른 프로젝트에 영향을 미치지 않게 할 수 있다. 가상환경을 사용하지 않으면, 시스템 전체에 패키지를 설치하게 되어 다른 프로젝트와 종속성이 엉킬 수 있다.
3. **일관된 개발 환경 유지**
- 프로젝트가 진행되는 동안 사용하는 패키지의 버전이 변경되지 않도록 할 수 있다. 특히, 팀 작업이나 협업 시 다른 개발자들과 동일한 환경을 유지하는 것이 중요하다. 가상 환경을 사용하면 특정 프로젝트에 맞는 패키지 버전을 유지하고 관리할 수 있다. 
4. **컴퓨터 용량 효율 관리**
- 가상환경을 사용하는 것이 전체 시스템에 설치하는 것 보다 오히려 더 작은 용량을 사용할 수 있다. 각 가상환경은 필요한 패키지만 설치하므로 불필요한 패키지 설치를 최소화할 수 있다. 또한 컴퓨터에 여러 버전의 Python과 패키지를 설치하고 관리하는 복잡성을 줄일 수 있다.

결론적으로, 가상환경은 **프로젝트별 종속성을 관리**하고 **패키지 충돌을 방지**하며, 유지보수를 편리하게 하기 위해 사용된다. 

가상환경을 쉽게 비유하자면 **각 프로젝트별로 폴더를 만들어 그 안에서 필요한 도구와 자원을 관리하는 것**과 비슷하다.

가상환경을 설정 하는 법은 다양한데,

Pycharm이나 VS code같은 대표적인 IDE 에서 가상환경을 설정하거나

터미널을 통해 가상환경을 켜고, Jupyter Notebook에서  연결하는 법 등이 있다.
