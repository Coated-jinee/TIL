# 팀 프로젝트
## 역할분배
나: 데이터 수집담당

채용 정보 오픈 API 소스 검색: 잡코리아, 사람인, 공공데이터 등

### 업무진행사항
사람인 API 키 신청 -> 승인(완료)
- 미리 신청한 팀원이 있었는데 신청 기간차이가 많이 안나기는 하지만 팀원은 아직 승인이 되지 않았다고 해서, 담당자에게 메일 보냄

공공 데이터 API: 사람인 API를 신청 중에 사용 할 테스트 데이터 서치
    - **서울시 일자리포털 채용 정보** 인증키 발급 (완료)
        https://data.seoul.go.kr/dataList/OA-13341/A/1/datasetView.do

만약 사람인 API 키 승인이 늦게 났으면, 지역별 공공데이터 채용 정보를 수집하려고 했음.
공공데이터는 행정 별로 구분이 되어있을 것 같아 행정구역 별 수집
일단은 큰 도시(광역시) 부터 ~ 소도시

**내일 오전에는 빠르게 아래 지역의 채용정보 공공 API를 발급받아보도록 하자** (뭐라도 더 있으면 좋음)
