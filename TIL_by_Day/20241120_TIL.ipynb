## **Embedding** 
왜 임베딩에서 벡터를 쓰는가? 
1. 복잡한 정보를 담기 위해
    - 스칼라는 단순히 하나의 값(예: 5, 10.2)으로 정보를 표현한다.
    - 하지만 자연어, 이미지, 사용자 데이터와 같은 복잡한 정보를 단일 값으로 표현하기에는 너무 제한적이다.
    - **벡터**는 여러 차원을 가지기 때문에, 다양한 속성을 동시에 표현할 수 있다.

2. 다차원 공간에서의 비교
    - 벡터는 **다차원 공간**에 위치하기 때문에 다른 벡터와의 유사도를 측정할 수 있다.
    - 예: 코사인 유사도(cosine similarity)로 두 단의의 의미적 유사성을 비교

3. 연산 가능성
    - 벡터는 수학적으로 더 많은 연산이 가능하다.
        - 예: 두 단어 벡터를 더하면 새로운 벡터를 얻을 수 있다.

스칼라로는 안 되는 이유
- 스칼라는 단일 값이라 의미의 복잡성과 관계를 표현하기 어렵다.
- 예를 들어, "고양이"와 "개"의 관계를 1차원 값으로는 표현하기 어렵지만, 벡터 공간에서는 둘의 거리를 계산하여 쉽게 비교할 수 있다. 

**왜 코사인 유사도를 쓰는가?**
**코사인 유사도**를 사용하는 이유는 벡터 간의 **방향성**을 비교하여 유사성을 측정하기 때문이다. 코사인 유사도는 벡터의 크기(길이)가 아니라 **방향**에 초점을 맞추기 때문에, 특히 자연어 처리(NLP)나 추천 시스템과 같은 작업에서 유용하다.
1. 크기의 영향을 제거
    - 벡터의 크기(길이)는 무시하고 방향만 비교
2. 유사도 측정에 적합
    - 벡터의 각도가 작을수록 유사도가 높다고 간주함
    - 특히, 임베딩 공간에서는 의미적으로 비슷한 데이터가 비슷한 방향을 가지는 경향이 있음
3. 고차원 데이터에서 효율적
    - 코사인 유사도는 고차원 데이터에서도 빠르고 효과적으로 유사도를 계산할 수 있음
    - NLP에서는 단어, 문장, 문서 벡터 간의 비교에 자주 사용된다.
4. 스케일링 문제 해결
    - 코사인 유사도는 벡터의 크기 차이를 무시하므로, 입력 데이터가 정규화되지 않아도 비교가 가능하다.
