### R 특강 D-day -1 

### 빔 프로젝터에 노트북을 연결할때는 해상도를 확인하자.
- 맥북과 빔프로젝터가 연결은 됐지만, 빔프로젝터에 맥북화면이 뜨지 않았다. 결국에 관제팀이 와서야 해결되었다. 원인은 의외로 단순했는데, 빔프로젝터가 지원하는 해상도나 주사율(50Hz)에 맞추지 않아서 화면이 정상적으로 표시되지 않는 것 이였다. 빔프로젝터는 특정 해상도와 주사율에서만 최적의 화면을 출력할 수 있는데, 설정이 맞지 않으면 빈 화면이 나오거나 화면이 깜빡이는 등 문제가 생길 수 있다.

해상도와 주사율 설정이 잘못 맞춰진 경우 빈화면 신호가 들어가거나, 지나치게 높은 해상도 신호가 빔프로젝터로 들어가 과부하가 발생할 수 있다.
