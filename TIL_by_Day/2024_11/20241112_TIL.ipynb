### R 특강 1일차

### 빔 프로젝터에 노트북을 연결할때는 해상도를 확인하자.
라고 어제의 소제목을 썼는데, 소제목을 정정한다.

### 빔 프로젝터에 노트북을 연결할 때는 (웬만하면) 윈도우 노트북을 쓰자.
강의실 환경이 나한테는 조금 복잡하긴 했다. 아직도 뭐가 어떻게 된건지 잘 모른다. 
어제 분명 빔프로젝터에 내 맥북이 알아서 미러링 되는걸 확인했고
오늘 똑같은 상황에서 연결했는데 이번엔 미러링이 안되고 화면분할이 됐다. 근데 디스플레이 설정창이 화면 분할이 된 곳으로 날아가버렸는데, 그 부분에 모니터가 없어서 디스플레이 설정을 못해..
관제팀 불렀다. 관제팀도 미러링을 해야한다고 했는데, 맥북은 익숙하지 않다고 하셔서(이해함..) 빈화면으로 날아간 설정창을 열심히 드래그로 찾았다.
선도 계속 뺐다 꽂았다 하면서 운좋게 설정창을 찾아서 해상도랑 주사율을 맞춰주셨다. 
그러고 교수님이랑 강의 교대하면서 교수님 맥북 꽂으니까 도돌이표. 

그래도 상황은 조금 나았던게, 검색으로 미러링 단축키가 있다는걸 알아냈다. (Command + F1)
근데 미러링은 해서 디스플레이 창을 띄웠는데 일단 디폴트 해상도에 50hz 선택이 없었다. (어제 50hz 맞추라는 것만 기억함)
해상도 옵션 열어서 몇개 다 바꿔봤는데 50hz짜리 옵션은 없었다.
이거 찾느라 점심시간도 반납했고, 교수님은 식사하러 가셨는데 노트북을 들고 가셔서 혼자 뭐 해볼수도 없었다. 
웃긴건 다시 내껄로 했는데도 안돼더라. (내가 안됐던 이유는 아마 이것저것 만지다가 HDMI2에서 1로 전환을 해버린듯 함. 
이때는 이 HDMI가 각 어디에 연결되어있는지도 제대로 몰랐음)

그래서 관제팀을 다시 호출하였다.
이번에는 교수님 HDMI 1과 2 




미러링 단축키가 있는 줄 몰랐다. (command + F1) 


HDMI1 - 윈도우 PC  -> 모니터가 확장식으로 빔프로젝터 연결 
HDMI2 - 빔프로젝터 연결

HDMI 리모콘이 따로 있고, 
윈도우 PC와 노트북을 전환하는 터치패널이 있고 여기에서는 빔프로젝트 1, 2 전환이 가능했다.




그리고 디스플레이 미러링 단축키가 있다는 걸 몰라서 디스플레이 설정 창이 빈화면(모니터 없음)에 화면분할로 들어갔다.
