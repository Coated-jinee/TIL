### Jira 소개 및 사용법

**Jira**는 Atlassian에서 개발한 **프로젝트 관리 및 이슈 추적 소프트웨어**로, 팀이 작업을 계획, 추적, 협업하는 데 사용된다.

## 1. Jira의 주요 목적
- **프로젝트 관리**: 작업 상태, 우선순위, 담당자를 체계적으로 관리.
- **이슈 관리**: 버그, 작업, 사용자 스토리 등 모든 작업 항목 관리.
- **협업 도구**: 팀원들과 실시간으로 진행 상황 공유.

---

## 2. Jira의 대표적인 기능

### 1) 보드 관리 (Kanban, Scrum)
- **Kanban 보드**: 작업 상태를 시각적으로 관리.
- **Scrum 보드**: 스프린트 기반의 작업 계획 및 진행.

### 2) 이슈 관리
- 이슈 유형: 
  - Bug (버그)
  - Story (사용자 스토리)
  - Task (작업)
  - Epic (큰 작업 단위)

### 3) 프로젝트 추적
- 작업 상태와 진행 상황을 시각화.
- 보고서 기능 예시:
  - 버그 해결 시간
  - 스프린트 속도 (Velocity)
  - 작업 완료 비율

### 4) 통합 기능
- GitHub, Bitbucket, Slack 등 다양한 툴과 연동 가능.

---

## 3. Jira에서 사용하는 주요 개념

### 1) 이슈(Issue)
- Jira에서 모든 작업 항목은 '이슈'로 관리.
- 예: 로그인 기능 개발, 버그 수정.

### 2) 에픽(Epic)
- 여러 이슈를 묶은 큰 작업 단위.
- 예: "결제 시스템 구축."

### 3) 스프린트(Sprint)
- 일정 기간(1~2주) 동안 작업을 완료하기 위한 계획.

### 4) 워크플로우(Workflow)
- 작업 상태 및 흐름 정의:
  - 예: To Do → In Progress → Done.

---

## 4. Jira 사용 사례
1. **소프트웨어 개발**:
   - 애자일 방식으로 프로젝트 계획 및 관리.
2. **버그 추적**:
   - 발견된 버그를 기록하고, 해결 과정 추적.
3. **비개발 팀의 업무 관리**:
   - 마케팅, 디자인, HR 등 다양한 팀에서 활용.

---

## 5. Jira와 GitHub 연결하기

Jira와 GitHub를 연동하면 코드 작업과 이슈를 통합 관리할 수 있다.

### 연동 방법 요약:
1. **Jira와 GitHub 계정 연결**:
   - Jira의 "앱 관리"에서 GitHub Marketplace를 통해 연동 앱 설치.
   - GitHub 리포지토리를 Jira 프로젝트와 연결.
2. **작업 추적**:
   - GitHub 커밋 메시지에 Jira 이슈 키를 포함하면 자동 연동.
     - 예: `git commit -m "Fix login bug [PROJECT-123]"`
3. **변경사항 확인**:
   - Jira 이슈에서 관련 GitHub Pull Request와 커밋 정보 확인.

### Jira를 통한 UI 용어 공부
- 여전히 UI 용어를 모르는게 너무 많음. 용어를 알아야 대화가 가능함.

1. 그림에서 Co와 Ju가 있는 부분 (워크스페이스 아이콘)
이 부분은 일반적으로 워크스페이스 스위처(Workspace Switcher) 또는 사이드바 네비게이션(Sidebar Nevigation)으로 불린다.
- Workspace Switcher: 워크 스페이스를 전환하는 버튼이 포함 된 영역
- Sideber: 하면 왼쪽에 있는 네비게이션 패널 전체를 의미
    - Jira에서는 워크스페이스 아이콘(Co, Ju)과 주요 앱 메뉴(Atlassian Home, Jiram Confluence 등)가 포함된 부분을 말함

2. Co와 Ju를 클릭했을 때 나오는 메뉴 (드롭다운 메뉴)
Co와 Ju를 클릭하면 나타나는 메뉴는 드롭다운 메뉴(Dropdown Menu) 또는 **워크스페이스 메뉴(Workspace Menu)**라고 부름
- Dropdown Menu: 클릭 시 펼쳐지는 선택 메뉴
- Workspace Menu: 워크스페이스 관련 옵션이 포함된 메뉴

3. Atlassian Home, Jira, Confluence 등이 표시된 부분
이 부분은 보통 메인 메뉴(Main Menu) 또는 앱 메뉴(App Menu)로 불린다.

4. "내 작업"과 같은 주요 탭 영역
상단의 "내 작업", "프로젝트", "필터", "대시보드" 등은 탭 메뉴(Tab Menu) 또는 탑 내비게이션(Top Navigation)으로 불린다.
- Top Navigation: 화면 상단의 주요 탐색 메뉴
- Tab Menu: 여러 탭으로 이루어진 메뉴

5. 사이드바 전체를 부르는 이름
화면 왼쪽에 있는 전체 영역(Co, Ju, Atlassian Home 등 포함)은 **사이드바(Sidebar)로 통칭된다.
