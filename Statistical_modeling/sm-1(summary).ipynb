### 통계모델링 1강

수업 대상 및 요구사항

	•	기초 수학적 지식: 고등학교 수학 범위 내에서 다루지만, 통계학 기초는 필요하다.
	•	주요 내용: 회귀분석, 수리통계학, 일반화선형모형(GLM) 기초 및 확장.
	•	베이즈 모형에서는 WinBUGS를 사용해 모수를 측정한다.
	•	예제 데이터 다운로드.

통계의 두 축

	•	추정과 검정이 통계의 두 가지 중요한 축이다.
	•	통계 모델링의 목적:
	•	설명력과 예측력을 갖춰야 한다.
	•	대표적인 예로 선형회귀가 두 요소를 잘 수행한다.

1.1 통계모형: 왜 통계적인 모형인가?

	•	통계모형은 데이터를 바탕으로 현상을 설명하는 도구이다.
	•	오차를 확률분포로 표현하여, 모델이 실제 데이터를 얼마나 잘 설명하는지 평가한다.

수식:  Y = M + \varepsilon 

	•	 Y : 관측된 데이터 (종속변수).
	•	 M : 모델로 설명되는 부분.
	•	 \varepsilon : 오차.
	•	가정이 맞지 않으면, 현상을 왜곡할 수 있다.

회귀분석의 현실

	•	회귀분석은 현장에서 자주 사용되지만, 정교하게 다루기는 쉽지 않다.
	•	데이터사이언티스트는 통계학적 원리를 알고 있어야 하며, 유연성과 정확성의 균형을 맞출 수 있어야 한다.

모델과 데이터의 적합성

	•	모델이 데이터에 얼마나 잘 맞는지 정량적으로 평가하는 절차가 필요하다.
	•	데이터 타입과 분포(가우시안, 감마, 포아송, t분포 등)에 대한 이해가 필수적이다.

오차의 종류

	•	조정 가능한 오차와 조정 불가능한 오차가 있으며, 조정 가능한 오차를 줄이는 것이 모델링의 목표다.

모델의 간결함

	•	간결한 모델이 설명력을 높일 수 있으며, 이는 통계 모델링에서 중요한 요소이다.

확률분포의 역할

	•	확률분포를 사용하면 오차와 결측 데이터를 효과적으로 표현할 수 있다.
