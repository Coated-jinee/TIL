#### 확률분포화 통계모형의 최우추정
확률분포(probability ...)는 통계모형의 본질적인 도구이고, 데이터에 보이는 다양한 오차(랜덤속성=확률적속성)를 표현한다.

#### 2.1 예제: 종자수의 통계모델링
식물 50개체로 구성된 집단에 대해서, 각 개체마다 종자(씨앗)수를 정리한 수치를 데이터라고 하겠다.
이 종자수 데이터를 통계모형으로 표현하게 위해 가장 어울리는 확률분포는 무엇일까요? 
우선 이 데이터는 0개, 1개, 2개 처럼 셀 수 있는 카운트 데이터 이다.

EDA = 탐색적 분석 = 90%의 signal을 알 고 간다.
기초통계량 알아보고 감. summary()

Quantile (4분위수)

IQR - interquantile range : 얼마나 퍼져있느냐 = 분산 . 분산은 데이터 정보량

데이터 분석에서 가장 중요한 것은, 일단은 무조건 데이터를 다양한 방법으로 도시(Visualization)해 보는 것

**카운트 데이터 나오면 : 포아송 분포를 이용**해서 모델링 이라고 외우기

포아송 분포의 정의가 단위시간(구간,거리)내에 발생하는 횟수 (포아송 확률변수)

확률분포란 확률변수(random variable)가 취할 수 있는 값과 그 값이 출현할 확률을 대응시킨 것이다.
확률 분포란, "이 값이 나올 가능성은 이정도야!"라고 알려주는 지도 같은 개념이다. 

확률 분포를 **사탕 뽑기 상자**로 비유할 수 있다.
상자 안에 다양한 색의 사탕이 들어있는데, 각 색이 나올 확률이 다르다고 가정해본다.
확률변수는 그 사탕의 색깔이 될 수 있고, 
확률은 각 색깔이 나올 가능성을 나타냅니다.
