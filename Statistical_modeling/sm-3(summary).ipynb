### 베이지안 모델링과 통계 용어

#### 통계학 용어 정리

	•	일탈도 (Deviance): 모델과 실제 데이터 간 차이를 측정하는 지표. 데이터를 얼마나 잘 설명하는지 나타낸다.
	•	잔차일탈도 (Residual Deviance): 모델이 데이터의 변동성을 얼마나 잘 설명하는지 확인할 수 있는 지표. 잔차의 일탈도를 나타낸다.
	•	최대일탈도 (Null Deviance): 아무 설명변수도 없는 가장 단순한 모델과 실제 데이터 간 차이를 나타낸다.
	•	고정효과 (Fixed Effects): 고정된 값에서 설명변수가 반응변수에 미치는 영향을 측정한다.
	•	랜덤효과 (Random Effects): 무작위로 선택된 값에서 설명변수가 반응변수에 미치는 영향을 측정한다.
	•	링크 함수 (Link Function): GLM에서 반응변수의 평균과 선형 예측 변수 간의 관계를 변환해주는 함수. 반응변수의 분포를 변형한다.
	•	반응변수 (Response Variable): 종속변수. 예측의 결과로 설명되는 변수이다.
	•	설명변수 (Explanatory Variable): 독립변수. 반응변수에 영향을 주는 변수이다.

#### 확률분포와 통계모형

	•	확률분포 (Probability Distribution): 확률변수가 가질 수 있는 값들과 그 값이 발생할 확률을 나타낸다.
	•	포아송 분포 (Poisson Distribution): 주어진 시간이나 구간 내에서 발생하는 사건의 횟수를 모델링하는 데 사용한다. 카운트 데이터 모델링에 적합하다.

#### 확률변수와 확률변수 값의 차이

	1.	확률변수 (Random Variable): 확률 실험의 결과를 수치로 표현하는 함수이다. 다양한 결과를 가질 수 있는 가능성을 나타낸다.
	2.	확률변수 값 (Value of a Random Variable): 특정 실험에서 확률변수가 취한 구체적인 값이다.

#### 핵심 차이
확률변수는 여러 값을 가질 수 있는 함수이고, 확률변수 값은 그 함수가 특정 상황에서 취한 하나의 결과이다.

#### 데이터 분석의 핵심

	•	데이터를 시각화하는 것이 매우 중요하다. 다양한 방법으로 데이터를 시각화해 데이터를 이해하고 패턴을 발견하는 것이 필요하다.
	•	**탐색적 데이터 분석 (EDA)**는 데이터 분석의 첫 단계로, 데이터를 요약하고 이해하는 데 집중한다. 이를 통해 데이터의 특성을 파악할 수 있다.
	•	기초 통계량을 확인하는 것은 데이터 분석의 기본이다.
	•	**IQR (Interquartile Range)**는 데이터의 분산 정도를 나타내는 지표로, 데이터가 얼마나 퍼져 있는지를 알 수 있다.
