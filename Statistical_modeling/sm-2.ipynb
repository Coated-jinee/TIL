#### 1.3 일반화선형모형(GLM)의 도입과 그 베이즈적인 확장

베이즈통계학을 쓰려면 일반화선형모형(generalized linear model, GLM)에 대한 이해가 있어야 함

선형모형의 발전 y = X\beta
    y: 결과 변수 (종속변수)로 여러 샘플의 결과 값을 나타내기 때문에, 여러 값을 가진 열 벡터일 수 있다.
    X: 설명 변수의 행렬로, 각 열은 하나의 독립 변수를 나타내고 각 행은 관측값을 나타낸다. 즉, 관측값들이 행을 이루고 각 열은 변수(특징)로 구성된 **행렬**이다.
    \beta: 회귀 계수들의 벡터로, 각 독립 변수에 대응하는 가중치(계수)를 나타낸다. 즉, 각 독립 변수에 대한 가중치를 모아 놓은 벡터이다. 

선형 모형의 데이터가 존재한다고 했을 때, 선형 모형의 직선을 특징화시켜주는 것은 기울기, 절편 이다. 절편과 기울기를 알면 선형 모형의 직선을 알 수 있다.

이 값을 추정하는 방법 중 대표하는 방법이 최소제곱법이 있다. 
선형모형은 정규분포를 가정하고 있다.(반응 변수)
예시: 온도와 아이스크림의 판매량(연속형 변수). 판매량이 연속형 변수가 아니고 count 데이터 일수 있는데(판매 갯수, 일사병의 발생 빈도) 카운터 변수면 정규분포 연속형 변수에 있는 기법을 쓸 수가 없게 된다. 쓸 수는 있는데 왜곡 된 결과 나옴.
그래서, 정규분포 이외의 카운터 데이터를 다루고 싶을 때는 일반화 선형모형을 확장시켜버린다. 

**일반화선형모형의 대표적인 확장**
반응변수에 대한 가정이 정규분포에 이 외 다른것으로 넘어간 것. 그래서 일반화임
(1)로지스틱 회귀모델 - 반응 변수가 이항분포. (예시: 있다/없다.)
(2)포아송 회귀분석 - Count. (예시: 씨앗을 터트릴 수 있는 정상적인 종자의 갯수)

count 변수인것을 일반적인 정규분포 적용시키면 결과는 나오는데 절망적임 (왜냐면? 틀린건데 결과가 나오니까 맞는줄 앎)

**일반화혼합모형** mixed model
반응변수의 가정이 다양한 경우 ..  
"데이터가 추출 된 개체간의 변수의 측정치는 서로 독립이어야 한다"가 기본 가정치인데, 독립이 아닐 때가 있음.
[예시]
Case 1- 혈압을 측정하는데, 100명의 환자가 있음. 그다음 혈압을 한명씩 측정하면 샘플이 100개가 나옴.
Case 2- 10명의 환자에게, 10번씩 측정 -> 샘플 100개 나옴 
이 경우 데이터 갯수는 100개, 100개로 동일. 이 결과를 기본 선형 모형으로 돌리면 결과 똑같이 나옴. 똑같이 나오지만 그대로 쓰면 안된다. 
'독립' = 서로 다른 개체로부터 출발해야한다. 

한 개체에서 여러번 하는 경우, 
독립성이 깨진 자료를 **반복측정자료**라고 하거나 **경시적 자료분석** 이라고도 함

Demographic data -> 인구 통계 정보
이런거 할때 묻지마 통계 분석 하면 데이터 짬뽕된다.

개체간에서 일어나는 분산을 따로 구별해서 고려하는 것을 혼합모형이라고 한다.
random effect(개체안에서 일어나는 랜덤성을 캐치해서 넣어줌) + fixed effect (기본적인 변수)

**계층베이즈모형**

논문 쓸때 추정모수 5개- 10개 사이임. 그 이상 못늘림 (너무 많아서 )
n은 작은 데 추정에 대한 모수가 큰 경우가 있음 n < p 이렇게 하면 일반적으로 회귀분석 안돌아감
(관측이 100개인데, 추정 모수가 50개면 이미 사이비임)

이 문제들(선형모형, 일반화선형모형, 일반화혼합모형에서 나오는)을 한방에 모아서 끝내버릴 수 있는게 베이지안 모델임

그러나 베이즈모형은 한가지 치명적인 단점이 있음. 컴퓨팅 리소스가 있어야 함. 컴퓨팅 리소스가 작으면 정밀하게 추적하기 어려움
컴퓨팅 리소스가 많아지고, 데이터가 많아지고 하니까 점점 나아지고 있음 


**"절대 진리":**통계학에서 "절대 진리"라는 개념은 고전적 빈도주의(Frequentist)접근과 베이지안(Bayesian)접근의 사고방식과 연관이 있다. 이 개념은 철학적인 사고방식에도 영향을 받으며, 이를 통해 통계적 추론 방식을 설명할 수 있다.   

1. 절대 진리와 빈도주의(Frequentist Approach):
    - 절대 진리는 세상에 존재하는 하나의 고정된 진리나, 측정하거나 알아낼 수 있는 하나의 '참된 값'을 의미한다. 
    - 빈도주의 통계에서는 어떤 사건의 확률을 장기적으로 반복되는 실험에서의 빈도로 정의한다. 즉, 모집단에는 하나의 고정된, 참된 값이 있고, 우리는 이를 추정하는 방식으로 접근한다. 
2. 포스트모더니즘과 베이지안(Bayesian Approach):
    - 포스트모터니즘적인 사고방식에서는 절대 진리가 존재하지 않는다고 본다. 진리는 고정된 것이 아니라, 개인의 관점이나 상황에 따라 변화한다고 생각한다. 이 사고방식은 베이지안 통계와 연결 될 수 있다.
    - 베이지안 통계에서는 확률을 어떤 사건이 발생할 가능성에 대한 개인의 신념으로 해석한다. 즉, 데이터와 새로운 정보다 주어지면 그에 따라 우리의 신념(혹은 확률)이 변화할 수있다는 것이다.
