2강 - 통계모델링의 전체조망

1.2 블랙박스 같은 통계분석의 악몽

통계 분석의 잘못된 사용 사례

	•	유의차가 나올 때까지 검정 방법을 계속 바꾼다.
	•	관측치들끼리 나눗셈을 해서 새로운 지표를 만들어 유의차가 나올 때까지 반복한다.
	•	R² 값은 설명력이므로 무조건 1에 가까우면 좋다고 생각한다.
	•	설명력은 분산 설명력과 모델 해석으로 나뉜다.
	•	R²는 변수를 추가할수록 커지지만, 모델의 설명력을 희석시킬 수 있다.
	•	등분산 가정이 위반되면 데이터를 변환해 회귀 분석이나 ANOVA를 시도한다.
	•	통계에서 가정이 깨졌을 때에도 사용할 수 있는 분석 방법들이 있다.
	•	다중검정을 하지 않았다고 지적받으면 무조건 보정을 하려고 한다.
	•	유의수준 0.05는 오류 5%를 허용하는데, 샘플 수가 많으면 절대적인 오류 수가 크게 증가한다.

통계학의 본질

	•	통계학은 Art적인 속성을 가진다. 즉, 교과서에 나오는 규칙을 넘어서는 가치 판단을 포함한다.
	•	가치 판단의 기준은 수리통계학에 근거한다. 이를 통해 오류와 왜곡을 잡아내는 것이 중요하다.

1.3 일반화선형모형(GLM)과 베이지안 모델의 확장

	•	베이지안 통계학을 사용하려면 **일반화선형모형(GLM)**에 대한 이해가 필수다.
	•	선형모형  y = X\beta 에서 반응변수  y 와 설명변수 행렬  X , 그리고 회귀 계수 벡터  \beta 로 구성된다.
	•	선형모형에서 정규분포를 가정하지만, 카운트 데이터와 같은 경우에는 일반화선형모형을 사용해야 한다.

일반화선형모형의 확장

	1.	로지스틱 회귀모델: 반응변수가 이항분포를 따를 때 사용 (예: 있다/없다).
	2.	포아송 회귀분석: 카운트 데이터에 적용 (예: 씨앗의 개수).

일반화혼합모형 (Mixed Model)

	•	기본 선형모형의 가정인 데이터의 독립성이 깨질 경우, 혼합모형을 사용한다.
	•	예시: 한 개체에서 여러 번 측정한 반복측정자료는 독립적이지 않다.
	•	혼합모형은 개체 간의 변동성(랜덤 효과)과 일반적인 변수(고정 효과)를 모두 고려한다.

계층 베이즈 모형

	•	추정해야 할 모수가 너무 많으면 일반적인 회귀 분석이 불가능할 수 있다.
	•	베이즈 모형은 이러한 문제들을 해결할 수 있지만, 컴퓨팅 리소스가 많이 필요하다.

절대 진리와 통계학

	1.	빈도주의(Frequentist): 하나의 고정된 진리가 존재하며, 이를 추정하려고 한다.
	2.	베이지안(Bayesian): 진리는 고정되지 않고, 개인의 관점이나 상황에 따라 변할 수 있다.
